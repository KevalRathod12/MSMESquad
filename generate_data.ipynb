```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import random
```

```python
# Configuration
DAYS_HISTORY = 365  # 1 year data
START_DATE = datetime.now() - timedelta(days=DAYS_HISTORY)

print("ðŸš€ Starting Data Factory for MSME Manufacturing...")
```

```python
# ---------------------------------------------------------
# 1. MACHINE SENSOR DATA (Predictive Maintenance)
# ---------------------------------------------------------
print("ðŸ›  Generating Machine Sensor Logs...")
machine_ids = ['CNC-001', 'CNC-002', 'PRESS-101', 'LATHE-55']
sensor_data = []

timestamps = pd.date_range(start=START_DATE, periods=DAYS_HISTORY * 24, freq='H') # Hourly data

for ts in timestamps:
    for mid in machine_ids:
        # Simulate normal vs abnormal behavior
        is_anomaly = random.random() < 0.02 # 2% chance of issue
        
        if is_anomaly:
            vibration = np.random.uniform(5.0, 8.5) # High vibration
            temperature = np.random.uniform(80, 110) # Overheating
            status = 'Risk'
        else:
            vibration = np.random.uniform(0.5, 3.0) # Normal
            temperature = np.random.uniform(40, 65) # Normal
            status = 'Normal'
            
        sensor_data.append([ts, mid, round(vibration, 2), round(temperature, 1), status])

df_sensors = pd.DataFrame(sensor_data, columns=['Timestamp', 'Machine_ID', 'Vibration', 'Temperature', 'Status'])
df_sensors.to_csv('machine_sensor_data.csv', index=False)
print("âœ… machine_sensor_data.csv created!")
```

```python

# ---------------------------------------------------------
# 2. INVENTORY & SALES DATA (Demand Forecasting)
# ---------------------------------------------------------
print("ðŸ“¦ Generating Inventory Sales Records...")
products = ['AutoPart-X', 'AutoPart-Y', 'Gear-Z', 'Bearing-B1']
sales_data = []

for day in range(DAYS_HISTORY):
    current_date = START_DATE + timedelta(days=day)
    # Weekend dip, seasonal spike logic
    is_weekend = current_date.weekday() >= 5
    season_factor = 1.2 if current_date.month in [10, 11, 12] else 1.0 # Diwali rush
    
    for pid in products:
        base_demand = 50 if pid == 'AutoPart-X' else 30
        daily_sales = int(np.random.normal(base_demand, 10) * season_factor)
        if is_weekend: daily_sales = int(daily_sales * 0.5)
        
        sales_data.append([current_date.date(), pid, max(0, daily_sales)])

df_inventory = pd.DataFrame(sales_data, columns=['Date', 'Product_ID', 'Quantity_Sold'])
df_inventory.to_csv('inventory_demand_data.csv', index=False)
print("âœ… inventory_demand_data.csv created!")

```

```python
# ---------------------------------------------------------
# 3. ENERGY CONSUMPTION DATA (Sustainability)
# ---------------------------------------------------------
print("âš¡ Generating Energy Usage Logs...")
energy_data = []

# Hourly data already in timestamps variable
for ts in timestamps:
    # Day vs Night consumption
    hour = ts.hour
    if 9 <= hour <= 18: # Shift time
        kwh = np.random.uniform(200, 350)
        load_type = 'Peak'
    else:
        kwh = np.random.uniform(50, 100)
        load_type = 'Off-Peak'
        
    energy_data.append([ts, kwh, load_type])

df_energy = pd.DataFrame(energy_data, columns=['Timestamp', 'kWh_Usage', 'Load_Type'])
df_energy.to_csv('energy_consumption_data.csv', index=False)
print("âœ… energy_consumption_data.csv created!")

```

```python
# ---------------------------------------------------------
# 4. QUALITY CONTROL MOCK DATA (Defect Detection)
# ---------------------------------------------------------
print("ðŸ” Generating Quality Control Logs...")
# Since we can't generate actual images, we simulate the 'Labels' for training
quality_data = []
for i in range(500):
    img_id = f"img_{i:04d}.jpg"
    is_defect = random.random() < 0.1 # 10% defect rate
    label = 'Defective' if is_defect else 'Good'
    defect_type = random.choice(['Crack', 'Rust', 'Dent']) if is_defect else 'None'
    
    quality_data.append([img_id, label, defect_type])

df_quality = pd.DataFrame(quality_data, columns=['Image_ID', 'Label', 'Defect_Type'])
df_quality.to_csv('quality_control_data.csv', index=False)
print("âœ… quality_control_data.csv created!")

print("\nðŸŽ‰ MISSION ACCOMPLISHED: All data files are ready for Step 2!")
```

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
from PIL import Image, ImageDraw, ImageFilter
import os
import random

# Folder Setup
TEST_FOLDER = 'Testing_Files'
if not os.path.exists(TEST_FOLDER):
    os.makedirs(TEST_FOLDER)

print("ðŸ§ª Generating REALISTIC Testing Data (with Noise & Textures)...")

# ==========================================
# 1. MAINTENANCE DATA (With Sensor Noise)
# ==========================================
print("ðŸ›  Creating 'test_maintenance.csv' with sensor noise...")
maint_data = []
machine_ids = [f"MACH-{i:03d}" for i in range(1, 21)]

for mid in machine_ids:
    is_risky = random.random() < 0.3
    
    # Base Values
    if is_risky:
        base_vib = random.uniform(6.0, 9.0)
        base_temp = random.uniform(90.0, 120.0)
    else:
        base_vib = random.uniform(0.5, 3.5)
        base_temp = random.uniform(40.0, 75.0)
    
    # ADDING NOISE (Realism)
    # Real sensors kadi sthir nathi hota, ema fluctuation hoy
    vib_noise = np.random.normal(0, 0.15) # Gaussian noise
    temp_noise = np.random.normal(0, 1.5)
    
    final_vib = round(max(0, base_vib + vib_noise), 2)
    final_temp = round(max(25, base_temp + temp_noise), 1) # Temp can't be below room temp
    
    maint_data.append([mid, final_vib, final_temp])

df_maint = pd.DataFrame(maint_data, columns=['Machine_ID', 'Vibration', 'Temperature'])
df_maint.to_csv(f'{TEST_FOLDER}/test_maintenance.csv', index=False)
print("âœ… Saved: test_maintenance.csv")

# ==========================================
# 2. INVENTORY DATA (With Demand Fluctuation)
# ==========================================
print("ðŸ“¦ Creating 'test_inventory_plan.csv'...")
future_dates = [datetime.now().date() + timedelta(days=i) for i in range(1, 31)]
products = ['AutoPart-X', 'AutoPart-Y', 'Gear-Z', 'Bearing-B1']

inv_data = []
for d in future_dates:
    # Weekend Logic (Noise in Demand)
    is_weekend = d.weekday() >= 5
    
    for p in products:
        inv_data.append([d, p])

df_inv = pd.DataFrame(inv_data, columns=['Date', 'Product_ID'])
df_inv.to_csv(f'{TEST_FOLDER}/test_inventory_plan.csv', index=False)
print("âœ… Saved: test_inventory_plan.csv")

# ==========================================
# 3. ENERGY DATA (With Load Spikes)
# ==========================================
print("âš¡ Creating 'test_energy.csv' with load spikes...")
energy_data = []
hours = [f"{i}:00" for i in range(24)]

for h in hours:
    hour_int = int(h.split(':')[0])
    
    # Base Load
    if 9 <= hour_int <= 18: # Shift Time
        base_load = random.randint(300, 480)
    else:
        base_load = random.randint(100, 200)
    
    # Add Noise (Machine startup spikes)
    spike = 0
    if random.random() < 0.1: # 10% chance of sudden spike
        spike = random.randint(50, 100)
        
    final_load = base_load + spike
    energy_data.append([h, final_load])

df_energy = pd.DataFrame(energy_data, columns=['Hour', 'kWh_Usage'])
df_energy.to_csv(f'{TEST_FOLDER}/test_energy.csv', index=False)
print("âœ… Saved: test_energy.csv")

# ==========================================
# 4. REALISTIC METALLIC IMAGES (Procedural Generation)
# ==========================================
print("ðŸ” Generating Realistic Metallic Surfaces...")

def generate_metal_texture(width, height):
    # Base Grey
    img = Image.new('RGB', (width, height), color=(180, 180, 180))
    draw = ImageDraw.Draw(img)
    
    # 1. Create Brushed Metal Look (Add thousands of random lines)
    for _ in range(2000):
        x = random.randint(0, width)
        y = random.randint(0, height)
        l_len = random.randint(5, 20)
        shade = random.randint(150, 210) # Varied grey shades
        draw.line((x, y, x + l_len, y + 2), fill=(shade, shade, shade), width=1)
        
    # 2. Add Grain/Noise (Dots)
    for _ in range(5000):
        x = random.randint(0, width-1)
        y = random.randint(0, height-1)
        noise_color = random.randint(100, 230)
        img.putpixel((x, y), (noise_color, noise_color, noise_color))
        
    return img

def draw_jagged_crack(draw, start_pos, end_pos):
    # Crack seedhi line na hoy, vanki chuki hoy
    x, y = start_pos
    end_x, end_y = end_pos
    
    current_x, current_y = x, y
    steps = 15
    
    step_x = (end_x - x) / steps
    step_y = (end_y - y) / steps
    
    for i in range(steps):
        # Random jitter
        next_x = current_x + step_x + random.randint(-3, 3)
        next_y = current_y + step_y + random.randint(-3, 3)
        
        # Draw thick dark line (Crack core)
        draw.line((current_x, current_y, next_x, next_y), fill=(20, 20, 20), width=3)
        # Draw thinner light line (Edge highlight)
        draw.line((current_x+1, current_y+1, next_x+1, next_y+1), fill=(100, 100, 100), width=1)
        
        current_x, current_y = next_x, next_y

def create_realistic_product(filename, is_defective):
    # 1. Get Metal Base
    img = generate_metal_texture(200, 200) # Thodi moti image
    draw = ImageDraw.Draw(img)
    
    if is_defective:
        # 2. Add Rust Patches (Katthai Dhabba)
        for _ in range(3):
            rx = random.randint(20, 150)
            ry = random.randint(20, 150)
            draw.ellipse((rx, ry, rx+30, ry+30), fill=(139, 69, 19, 100), outline=None)
        
        # 3. Add Jagged Crack
        draw_jagged_crack(draw, (50, 50), (150, 150))
        
        # Blur thodu karvanu jethi fake na lage
        img = img.filter(ImageFilter.GaussianBlur(radius=0.5))
        
    else:
        # Good product - Shiny finish highlight
        draw.line((0, 0, 200, 200), fill=(220, 220, 220), width=1)
        img = img.filter(ImageFilter.SHARPEN)

    img.save(f'{TEST_FOLDER}/{filename}')

# Generate High Quality Images
create_realistic_product('test_product_BAD.png', True)
create_realistic_product('test_product_GOOD.png', False)

print("âœ… Saved: Realistic Images (Check 'test_product_BAD.png')")
print(f"\nðŸŽ‰ ALL DATA REGENERATED! Now upload these files in your Dashboard.")
```
